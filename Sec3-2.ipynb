(3) Categorizing different types of operators

````python
queryForms = ['ask', 'select', 'describe', 'construct']
Patterns = ['path', 'group', 'optional', 'union', 'graph', 'bind', 'having', 'minus', 'filter', 'agg']
Modifiers = ['orderby', 'projection', 'distinct', 'reduced', 'offset', 'limit']
fn = ['fn', 'builtin']
other = ['values', 'service']

def countOpChangePair(df):
    df_queryForm = DataFrame()
    df_pattern = DataFrame()
    df_modi = DataFrame()
    df_fn = DataFrame()
    df_other = DataFrame()
    df_list = [df_queryForm, df_pattern, df_modi, df_fn, df_other]
    op_list = [queryForms, Patterns, Modifiers, fn, other]
    for i in range(len(df)):
        idx = df.index[i]
        res = [False for _ in range(len(op_list)*2)]
        for idx_op, opi in enumerate(idx):
            for ii, op_listi in enumerate(op_list):
                for ith, opii in enumerate(op_listi):
                    if opii in opi.lower():
                        res[idx_op*len(op_list)+ii] = True
                        break
        for ith in range(len(op_list)):
              if res[ith] and res[len(op_list)+ith]:
                    df_list[ith] = df_list[ith].append(df.iloc[i])
                    # from ipdb import set_trace; set_trace()
    return df_list
````

(4) Deletion of a single operator

````python
def extractFeatures(df):
    df_form = DataFrame()
    df_pattern = DataFrame()
    df_modi = DataFrame()
    df_fn = DataFrame()
    df_other = DataFrame()
    df_list = [df_form, df_pattern, df_modi, df_fn, df_other]
    op_list = [queryForms, Patterns, Modifiers, fn, other]
    for i in range(len(df)):
        idx = df.index[i]
        for ii, op_listi in enumerate(op_list):
            for ith, opii in enumerate(op_listi):
                if opii in idx.lower():
                    df_list[ii] = df_list[ii].append(df.iloc[i])
                    continue
    return df_list  
````

````python
df = pd.read_csv('results/operator/compUsedOperator_10_new.csv', index_col=0)
op_comp_single_type = extractFeatures(df)
name = ['queryForm', 'pattern', 'modifier', 'fn', 'other']
for i, dfi in enumerate(op_comp_single_type):
    dfi.to_csv(f'results/operator/{name[i]}_CompSingleOp_10.csv')
````

````python
df = pd.read_csv('results/operator/dataset-operator-count.csv', index_col=0)
df_list_lsq1 = extractFeatures(df)
name = ['queryForm', 'pattern', 'modifier', 'fn', 'other']
for i, dfi in enumerate(df_list_lsq1):
    dfi.to_csv(f'results/operator/{name[i]}_singleop_10.csv')
````

df_list_lsq1 is a list, ordered by different operator types, storing the number of occurrences of each operator in the corresponding dataset

op_comp_single_type is also a list ordered by different operator types, storing the number of occurrences of each operator's deletion in the corresponding dataset

````python
count_all = []
for i in range(len(df_list_lsq1)):
    idx = df_list_lsq1[i]['all'].index
    temp = {}
    for j, idxi in enumerate(idx):
        temp[idxi] = df_list_lsq1[i]['all'][j]
    count_all.append(temp)

op_rewrite = []
for op_typei in range(len(op_comp_single_type)):
    op_change = op_comp_single_type[op_typei]['all']
    index = op_comp_single_type[op_typei]['all'].index
    temp = DataFrame()
    temp['#change'] = op_change
    all_ = []
    for j, idxi in enumerate(index):
        # op = idxi.split('_')[0]
        op = idxi.split('_')[0:-1]
        opstr = ''
        for kk, stri in enumerate(op):
            if kk != 0:
                opstr += '_'
            opstr += stri
        op = opstr
        if op in count_all[op_typei].keys():
            all_.append(count_all[op_typei][op])
        else:
            all_.append(-1)
    temp['#all'] = all_
    op_rewrite.append(temp)
````

````python
name = ['queryForm', 'pattern', 'modifier', 'fn', 'other']
for i, dfi in enumerate(op_rewrite):
    dfi['perc'] = dfi['#change']/dfi['#all']
    dfi.to_csv(f'results/operator/{name[i]}_operator_collected.csv')
````

op_rewrite is a list that stores the deletions and addtions of different types of operators separately

````python
op_rewrite[0]
````

````python
pattern = op_rewrite[1]
# all, deletion, insertion
res = {}
op2idx = {'add': 2, 'delete': 1}
index = pattern.index
for i in range(len(pattern)):
    idx = index[i]
    op = idx.split('_')[0:-1]
    opstr = ''
    for stri in op:
        opstr += stri
    op = opstr
    ad = idx.split('_')[-1]
    if op not in res.keys():
        res[op] = [0,0,0]
    res[op][0] = pattern.iloc[i]['#all']
    res[op][op2idx[ad]] = pattern.iloc[i]['#change']
````

````python
pattern_df = DataFrame.from_dict(res, orient='index', columns=['#all', '#deletion', '#insertion'])
pattern_df
````
